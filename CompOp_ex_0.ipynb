```python
print('0-3: For and While')
n = 100
rand = np.random.rand(n)
index = np.arange(n)

A = datetime.now()

localSum = 0
for i in np.arange(n):
    localSum = localSum + rand[i]

B = datetime.now()

sum = np.sum(rand)

C = datetime.now()

print('time : for loop', (B - A).total_seconds())
print('time : numpy sum', (C - B).total_seconds())
print('sum slow: ' ,localSum)
print('sum fast: ', sum)


resultVector = np.zeros(n)
resultVector2 = np.zeros(0)

resultVector[0] = rand[0]
for i in np.arange(1,n):
    resultVector[i] = rand[i] + resultVector[i-1]

    
    
cumulativeSum = 0
for val in rand:
    cumulativeSum += val
    resultVector2 = np.append(resultVector2,cumulativeSum)
    
print('Check if arrays are equal ',np.array_equal(resultVector,resultVector2))
# print('random ', randomVector)
# print('result ', resultVector)
# print('result2 ', resultVector2)
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(index, resultVector)

ax[0].set(xlabel='index', ylabel='Value',
       title='accumulated random number')
ax[0].grid()

ax[1].plot(index, rand)

ax[1].set(xlabel='index', ylabel='Value',
       title='random number')
ax[1].grid()

fig.show()
print('For more plot variants: see https://matplotlib.org/stable/gallery/index.html')
```

```python
print('0-4: Arrays')
n = 10
arr = np.random.rand(n,n)
#Simple test to see if code is correct: the average of a array of ones should be 1!
#arr = np.full((n, n), 1.0)

res = np.zeros(shape = (n,n))

x = np.linspace(0, 1, n)
y = np.linspace(0,1, n)
X, Y = np.meshgrid(x, y)




def averageSlow(arr):
    returnArray = np.zeros(np.shape(arr))
    for i in np.arange(1,np.shape(arr)[0]-1):        
        for j in np.arange(1,np.shape(arr)[1]-1):
            returnArray[i,j]=(arr[i,j] + arr[i,j+1] + arr[i+1,j] \
                            +arr[i-1,j] + arr[i,j-1] + arr[i+1,j+1] \
                            +arr[i+1,j-1] + arr[i-1,j-1] + arr[i-1,j+1] )/9.0

    j=0
    for i in np.arange(1,np.shape(arr)[0]-1):  
        returnArray[i,j]=(arr[i,j] +arr[i-1,j] + arr[i+1,j] \
                             + arr[i,j+1] + arr[i+1,j+1] + arr[i-1,j+1] )/6.0
                              
    i= 0
    for j in np.arange(1,np.shape(arr)[1]-1):  
        returnArray[i,j]=(arr[i,j] + arr[i,j+1] + arr[i,j-1] \
                            +arr[i+1,j]  + arr[i+1,j-1]  + arr[i+1,j+1] )/6.0
                             
    j=-1
    for i in np.arange(1,np.shape(arr)[0]-1):  
        returnArray[i,j]=(arr[i,j] +arr[i-1,j] + arr[i+1,j] \
                             + arr[i,j-1] + arr[i+1,j-1] + arr[i-1,j-1] )/6.0
                              
    i= -1
    for j in np.arange(1,np.shape(arr)[1]-1):  
        returnArray[i,j]=(arr[i,j] + arr[i,j+1] + arr[i,j-1] \
                            +arr[i-1,j]  + arr[i-1,j-1]  + arr[i-1,j+1] )/6.0
    returnArray[0,0] = (arr[0,0] + arr[0,1] + arr[1,0]  +arr[1,1])/4.0
    returnArray[0,-1] = (arr[0,-1] + arr[0,-2] + arr[1,-1]  +arr[1,-2])/4.0
    returnArray[-1,0] = (arr[-1,0] + arr[-2,0] + arr[-1,1]  +arr[-2,1])/4.0
    returnArray[-1,-1] = (arr[-1,-1] + arr[-2,-2] + arr[-2,-1]  +arr[-1,-2])/4.0
    return   returnArray      

def averageRoll(arr):
    res =   (np.roll(arr,-1, axis=0)+ \
         np.roll(arr,-1, axis=1)+ \
         np.roll(arr, 1, axis=0)+ \
         np.roll(arr, 1, axis=1)+ \
         np.roll(arr, [1, 1], axis=(0, 1))+ \
         np.roll(arr,[ -1, 1], axis=(0, 1))+ \
         np.roll(arr,[1,-1], axis=(0, 1))+ \
         np.roll(arr,[-1,-1], axis=(0, 1))+ \
         np.roll(arr,-1, axis=1)) / 9.0
    return res
  
    


def averagePad(arr):
    #print(arr)
    res =  (arr  \
           +np.pad(arr,((0,0),(1,0)), mode='constant')[:, :-1] \
           +np.pad(arr,((0,0),(0,1)), mode='constant')[:, 1:] \
           +np.pad(arr,((1,0),(0,0)), mode='constant')[:-1, :] \
           +np.pad(arr,((0,1),(0,0)), mode='constant')[1:, :] \
           +np.pad(arr,((1,0),(1,0)), mode='constant')[:-1, :-1] \
           +np.pad(arr,((1,0),(0,1)), mode='constant')[:-1, 1:] \
           +np.pad(arr,((0,1),(1,0)), mode='constant')[1:, :-1] \
           +np.pad(arr,((0,1),(0,1)), mode='constant')[1:, 1:] )
    res[1:-1,1:-1]/=9.0
    res[1:-1,0]/=6.0
    res[1:-1,-1]/=6.0
    res[0,1:-1]/=6.0
    res[-1,1:-1]/=6.0
    res[-1,-1]/=4.0
    res[0,-1]/=4.0
    res[-1,0]/=4.0
    res[0,0]/=4.0
    return res
    
date_before = datetime.now()

avRoll = averageRoll(arr)

date_after = datetime.now()

print('Roll version', (date_after - date_before).total_seconds())   
date_before = datetime.now()

avPad = averagePad(arr)

date_after = datetime.now()

print('Pad version', (date_after - date_before).total_seconds())  

date_before = datetime.now()

avForLoop = averageSlow(arr)

date_after = datetime.now()
print('Slow version', (date_after - date_before).total_seconds())  

print('For more plot variants: see https://matplotlib.org/stable/gallery/index.html')



fig, ax = plt.subplots(nrows=int(1), ncols=int(3), figsize=(15,5))


ax[0].imshow(avRoll)
ax[1].imshow(avPad)
ax[2].imshow(avForLoop)




fig, ax = plt.subplots(nrows=int(1), ncols=int(3), figsize=(15,5))

ax[0].contourf(X,Y,avRoll)
ax[1].contourf(X,Y,avPad)
ax[2].contourf(X,Y,avForLoop)
```
