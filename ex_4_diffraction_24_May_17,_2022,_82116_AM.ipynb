```python
import numpy as np
import matplotlib.pyplot as plt
from random import gauss
from gausslet_decomposition import GAUSSLET
from diffraction_fraunhofer import DIFFRACTION_FRAUNHOFER
from diffraction_fresnel import DIFFRACTION_FRESNEL
```

```python

distance = 0.625*1e-1 * 4 
slitwidth = 0.1
plot_range_factor = 0.4

dx_all = 0.0001


# gausslet = GAUSSLET(slit = slitwidth, gridp = 501, wavelength =1, nrGausslet = 91, propagationDistance = distance, plot_domain_factor = 250, steps_to_plot= 5)
# gausslet.calcGausslet()


diff_fraunhofer = DIFFRACTION_FRAUNHOFER(slit = slitwidth, dx = dx_all, wavelength = 1e-3,propDistance = distance , plot_range = plot_range_factor, steps_to_plot = 5)
diff_fraunhofer.setupField()
diff_fraunhofer.analytical()
diff_fraunhofer.fraunhofer()


diff_fren = DIFFRACTION_FRESNEL(slit = slitwidth, dx = dx_all, wavelength = 1e-3, propDistance = distance , plot_range = plot_range_factor, steps_to_plot = 5)
diff_fren.setupField()
diff_fren.fresnel()



varFraun = diff_fraunhofer.IrradFFT/np.max(diff_fraunhofer.IrradFFT)
varFren = diff_fren.IrradFFT / np.max(diff_fren.IrradFFT)
# print(diff_fren.w==diff_fraunhofer.IrradTheory)

plt.plot(diff_fren.xPrime, varFren,'--', diff_fraunhofer.xPrime, diff_fraunhofer.IrradTheory/np.max(diff_fraunhofer.IrradTheory),diff_fraunhofer.xPrime,varFraun, markersize=2)
#plt.plot(diff_fren.xPrime, varFren,'--', diff_fren.xPrime, diff_fren.IrradTheory/np.max(diff_fren.IrradTheory),diff_fraunhofer.xPrime,varFraun, markersize=2)

#plt.plot(diff_fren.xPrime, varFren)#, diff_fraunhofer.xPrime, diff_fraunhofer.IrradTheory/np.max(diff_fraunhofer.IrradTheory),diff_fraunhofer.xPrime,varFraun)
#plt.xlim((-distance*50, distance*50))
#plt.plot(diff_fraunhofer.IrradTheory/np.max(diff_fraunhofer.IrradTheory))
#plt.plot(diff_fren.xPrime, var)
#plt.xlim((-distance*1, distance*1 ))
#plt.xlim((-distance*5, distance*5 ))
#plt.xlim((-distance*5,distance*5))

plt.xlim((-distance*0.5 , distance*0.5 ))
plt.legend(['fresnel','analytic', 'fraunhofer'])
plt.show()
```
