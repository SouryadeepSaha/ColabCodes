```python
from transferMatrix import TRANSFERMATRIX
import numpy as np
import matplotlib.pyplot as plt
```

```python
gridpoints = 101
wavelength = 4e-4
frequency = 3e+8/wavelength
nSiN = 2.0
nSiO2 = 1.5
dSiN = wavelength / (4*nSiN)
dSiO2 = wavelength / (4*nSiO2)

transfer = TRANSFERMATRIX( numberLayers = 101, refractiveIndex1 = 1.7445, refractiveIndex2 = 1.8, fMin = frequency * 0.9, fMax = frequency * 1.1, plotPoints = gridpoints, layerThickness1 = 4e-7, layerThickness2 = 4e-7)

transfer.setUpLayers()
transfer.runSimulation()
```

```python
transfer = TRANSFERMATRIX( numberLayers = 3, refractiveIndex1 = 1.7445, refractiveIndex2 = 1.85, fMin = 1e12, fMax = 1000e12, plotPoints = gridpoints, layerThickness1 = 4e-7, layerThickness2 = 4e-7)

transfer.setUpLayers()
transfer.runSimulation()
```

```python
transfer = TRANSFERMATRIX( numberLayers = 51, refractiveIndex1 = nSiO2, refractiveIndex2 = nSiN, fMin = frequency*0.8, fMax = frequency*1.2, plotPoints = gridpoints, layerThickness1 = dSiO2, layerThickness2 = dSiN)

transfer.setUpLayers()
transfer.runSimulation()
```
