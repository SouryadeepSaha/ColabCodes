```python
import sys
import os
#sys.path.append(os.path.abspath("content/ColabCodes"))
from gauss_seidel import GaussSeidel
from datetime import datetime
# from transferMatrix import TRANSFERMATRIX

import numpy as np
import matplotlib.pyplot as plt
# from random import gauss
#from fdtd_staggered import FDTD
# from fdtd_mur import FDTD_MUR
# from gausslet_decomposition import GAUSSLET
```

```python
h_grid = 0.01
x_len = 1.0
y_len = 1.0
n_x = x_len/h_grid
n_y = y_len/h_grid
i_iter = 1000
gauss_iteration = GaussSeidel( h_grid, x_len, y_len, i_iter)
```

```python

runBegin = datetime.now()
gauss_iteration.run()
runEnd = datetime.now()
u_forLoop = gauss_iteration.u_h
print('Run Time: ', (runEnd - runBegin).total_seconds())

gauss_iteration.plot()

```

```python
runSliceBegin = datetime.now()
gauss_iteration.runSlice()
# gauss_iteration.plotLaplce()
runSliceEnd = datetime.now()
print('Run Time: ', (runSliceEnd - runSliceBegin).total_seconds())
gauss_iteration.plot()
u_slice = gauss_iteration.u_h
```

```python
diff_error = np.sqrt(np.divide(np.sum(np.square(u_forLoop.reshape(int(n_y+1),int(n_x+1))-u_slice)),(n_y+1)*(n_x+1)))
print(diff_error)
```

```python
plt.imshow(u_forLoop.reshape(int(n_y+1),int(n_x+1))-u_slice)
plt.show()
```
